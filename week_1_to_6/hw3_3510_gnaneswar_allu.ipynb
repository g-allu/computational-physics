```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-10, 10, 0.01)

def f(x):
    return np.sin(4*x) * np.cos(4*x)

def g(x):
    return np.where(x==0, 1.0, np.sin(x)/x)
    

def h(x):
    arr = np.tan(2*x)
    arr[np.abs(arr)>20] = np.nan
    return arr

plt.figure(figsize=(10, 6))

plt.plot(t, f(t), label = "f(t) = sin(4t)cos(4t)")

plt.plot(t, g(t), label = "g(t) = sin(t)/t")

plt.plot(t, h(t), label = "h(t) = tan(2t)")

asym = (np.pi / 4) + (np.arange(-10,10) * np.pi / 2)

for p in asym:
    plt.axvline(x=p, linestyle='--')

plt.xlim(-10, 10)

plt.ylim(-5, 5)

plt.xlabel("Time (t)")

plt.ylabel("ouput")

plt.title("f(t), g(t), and h(t)")

plt.legend()

plt.show()
```
