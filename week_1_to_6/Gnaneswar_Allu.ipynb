Prime Numbers from 1 to 100

```python
for i in range(1,101):
    count = 0
    for j in range(1,i+1):
        if i%j==0:
            count +=1
    if count == 2:
        print(i)
```

Equation 1: x^2 (x=5, h=1,0.1,0.01)

```python
x=5
h1= 1
h2=0.1
h3=0.01
ans1 = ((x+h1)**2-x**2)/h1
ans2 = ((x+h2)**2-x**2)/h2
ans3 = ((x+h3)**2-x**2)/h3

print("using h=1: ",ans1,"| using h=0.1: ",ans2,"| using h=0.01: ",ans3)
```

Equation 2: x^3 + x^2 (x=5, h=1,0.1,0.01)

```python
x=5
h1= 1
h2=0.1
h3=0.01
ans1 = (((x+h1)**3+(x+h1)**2)-((x)**3+(x)**2))/h1
ans2 = (((x+h2)**3+(x+h2)**2)-((x)**3+(x)**2))/h2
ans3 = (((x+h3)**3+(x+h3)**2)-((x)**3+(x)**2))/h3

print("using h=1: ",ans1,"| using h=0.1: ",ans2,"| using h=0.01: ",ans3)
```

Equation 3: sin(x) (x=pi, h=1,h=0.1,h=0.01)

```python
import numpy as np
import math as math

x= math.pi
h1= 1
h2=0.1
h3=0.01
ans1 = (np.sin(x)-np.sin(x-h1))/h1
ans2 = (np.sin(x)-np.sin(x-h2))/h2
ans3 = (np.sin(x)-np.sin(x-h3))/h3

print("using h=1: ",ans1,"| using h=0.1: ",ans2,"| using h=0.01: ",ans3)
```

As the h gets smaller and smaller the value gets closer and closer to the true derivative of the function.
