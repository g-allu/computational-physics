```python
import matplotlib.pyplot as plt
import numpy as np

G=0.074e-11
m1=5.0
m2=10.0

r=np.arange(0.1,10.0,0.1)

F = G * m1 * m2 / r**2

plt.figure()

plt.plot(r, F)

plt.xlabel("Separation distance r (m)")
plt.ylabel("Graviational force magnitude F (N)")
plt.title("Nwetonion Gravitational Force vs Distance")
plt.show()
```

```python
import numpy as np
epsilon = 4e-4
sigma  = 6.0
rmin =5.5
rmax =10
dr = 0.001

ulj_list = []

r_range = np.arange(rmin,rmax, dr)

for r in np.arange(rmin,rmax, dr):
    ulj = 4.0 * epsilon * ((sigma/r)**12-(sigma/r)**6)
    ulj_list.append(ulj)

import matplotlib.pyplot as plt
plt.plot(r_range, ulj_list)
plt.show()
```
